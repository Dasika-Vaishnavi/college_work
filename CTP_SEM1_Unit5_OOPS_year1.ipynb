It serves as the constructor. It is automatically executed when an object of the class is created. This method is used generally to initialize the instance variables, although any code can be written inside it. 


### __del__() Method (Destructor)
The __del__() method is a special method inside a class. It serves as the destructor. It is executed
automatically when an object is going to be destroyed. We can explicitly make the __del__() method to
execute by deleting an object using del keyword. 


```python
class student:
  branch = "CSE"
  def __init__(self, rno, n):
    self.regno = rno
    self.name = n
  def __del__(self):
    print("this object is going to be deleted")
  def print_details(self):
    print(self.regno)
    print(self.name)
  s = student(536, "Vaishnavi")
  s.print_details()
```

### Static Methods 
A static method is a method marked with staticmethod decorator. Static methods are the methods that are
bound to a class rather than its object. They do not require a class instance creation. So, they are not
dependent on the state of the object. Static method does not have any additional arguments like self and
cls.

```python
class box:
  def __init__(self, l, b):
    self.length = l
    self.breadth = b
  def print_dimensions(self):
    print(self.length)
    print(self.length * self.breadth)
b = box(3, 4)
b.print_dimensions()
```

Creating a new class from existing class is known as inheritance.
###### Parent Class (or Base Class): The class from which features are inherited is called Base class.
Child Class (or Derived Class or Sub Class): The class into which features are derived is called
Derived class. 

```python
class person:
  def __init__(self, fname, sname):
    self.first = fname
    self.second = sname
  def display(self):
    print(self.first, self.second)
#the class student inherits from class person
#person is the parent class, student is the derived class
class student(person):
  def __init__(self, fname, sname, rno):
    super().__init__(fname, sname)
    self.regno = rno
  def display_student(self):
    print(self.first, self.second, "has rno", self.regno)

p = person("Rama", "Raju")
p.display()
s = student("Vaishnavi", "Dasika", 536)
s.display()
s.display_student()

```

### Method Overriding 
In method overriding, a base class and the derived class will have a method with the same signature. The
derived class method can provide different functionality when compared to the base class method.
1. Class having multiple methods with the same name, but with different arguments.
2. Python does not support method overloading by default.
3. The problem with method overloading in Python is that we may overload the methods but can
only use the latest defined method. 


```python
class person:
  def __init__(self, fname, sname):
    self.first = fname
    self.second = sname
  def display(self):
    print(self.first, self.second)
class student(person):
  def __init__(self, fname, sname, rno):
    super().__init__(fname, sname)
    self.regno = rno
  #here display name is used again; the name is over written
  #this is called method overriding
  def display(self):
    print(self.first, self.second, "has rno", self.regno)

p = person("Rama", "Raju")
p.display()
s = student("Vaishnavi", "Dasika", "536")
s.display()
   
```

### Operator Overloading
Operator Overloading means giving extended meaning beyond their predefined operational meaning.
When we use an operator on user defined data types, then automatically a special function or magic
function associated with that operator is invoked. Operators work according to that behaviour defined in
the magic functions. Changing this magic method’s code, we can give extra meaning to the operator. 


```python
#operator "+"
#magic function name "__add()__"
class complex:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def __add__(self, other):
    return self.a + other.a, self.b + other.b
o1 = complex(1, 2)
o2 = complex(3, 4)
print(o1+o2)
```

```python
#try and except block
nom = int(input())
denom = int(input())
try:
  quotient = nom/denom
  print(quotient)
except ZeroDivisionError:
  print("denominator != 0")
```

```python
#multiple except blocks
nom = int(input())
denom = int(input())
try:
  quotient = nom/denom
  print(quotient)
except NameError:
  print("Name Error is caused")
except ValueError:
  print("Value Error is caused")
except ZeroDivisionError:
  print("You are dividing with 0")
```

```python
#multiple exceptions in single block
nom = int(input())
denom = int(input())
try:
  quotient = nom/denom
  print(quotient)
except (NameError, ZeroValueError, ValueError, TypeError):
  print("please check")
```

```python
#else clause
#the try and except blocks can be followed by an optional else block which 
#executes when all except can satiate
try:
  print("Hello")
except:
  print("something is wrong")
else:
  print("good to go")
```
